```python
# 6) Agrupe os dados do DataFrame criado no Exercício 1 por faixas de idade (por
# exemplo, 20-29, 30-39, etc.) e calcule a média do salário para cada faixa de idade.
```

```python
faixasIdade = pd.cut(df['Idade'], bins=[20, 29, 39, 49], labels=['20 a 29', '30 a 39', '40 a 49'])
df['Faixa de Idade'] = faixasIdade
mediaSalarialIdade = df.groupby('Faixa de Idade')['Salário'].mean()
mediaSalarialIdade
```
